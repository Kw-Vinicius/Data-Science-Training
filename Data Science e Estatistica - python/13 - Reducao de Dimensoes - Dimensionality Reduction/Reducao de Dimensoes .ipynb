# Principal Component Analysis (PCA)
An√°lise dos Principais Componentes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

```python
# Iris
iris = datasets.load_iris()
```

```python
type(iris)
```

```python
print(iris['DESCR'])
```

```python
X = iris['data']
y = iris['target']
```

```python
X.shape, y.shape
```

```python
X = StandardScaler().fit_transform(X)
```

```python
iris['feature_names']
```

```python
pd.DataFrame(X,columns=iris['feature_names']).head()
```

```python
pca = PCA(n_components=2)
```

```python
pc = pca.fit_transform(X)
```

```python
pc.shape
```

```python
pc[:,1]
```

```python
y
```

```python
pc[y==0,1]
```

```python
plt.scatter(pc[:,0],pc[:,1])
plt.show()
```

```python
plt.scatter(pc[y==0,0],pc[y==0,1],label='Setosa')
plt.scatter(pc[y==1,0],pc[y==1,1],label='Versiclour')
plt.scatter(pc[y==2,0],pc[y==2,1],label='Virginica')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
# Tirando 2 dimens√µes , ainda temos 95% dos dados explicados
```

# D√≠gitos

```python
digitos = datasets.load_digits()
X = digitos['data']
y = digitos['target']
```

```python
X.shape, y.shape
```

```python
pca = PCA(n_components=2)
X_reduzido = pca.fit_transform(X)

print(f'Projetando {X.shape[1]} dimens√µes em apenas 2D')

plt.figure(figsize=(12,10))
plt.scatter(X_reduzido[:,0],X_reduzido[:,1],c=y,
           edgecolor='black',alpha=0.6,s=80,
           cmap=plt.cm.get_cmap('nipy_spectral',10))
plt.colorbar()
plt.show()
```

# TSNE

```python
# mais detalhado , por√©m muito mais devagar
from sklearn.manifold import TSNE
tsne = TSNE(random_state=123)

X_tsne = tsne.fit_transform(X) 
```

```python
X_tsne.shape
```

```python
plt.figure(figsize=(12,10))
plt.scatter(X_tsne[:,0],X_tsne[:,1],c=y,
           edgecolor='black',alpha=0.6,s=80,
           cmap=plt.cm.get_cmap('nipy_spectral',10))
plt.colorbar()
plt.show()
```

```python
X.shape
```

```python
pca = PCA(0.95)
```

```python
pca.fit_transform(X)
```

```python
pca.n_components_
```

```python
np.arange(1,pca.n_components_+1)
```

```python
plt.figure(figsize=(8,4))
plt.bar(np.arange(1,pca.n_components_+1),pca.explained_variance_ratio_[:pca.n_components_],
       color='lightskyblue')
plt.xlabel('Componentes Principais')
plt.ylabel('Varia√ß√£o')
plt.xticks(np.arange(1,pca.n_components_+1))
plt.show()
```
